```python
import tkinter as tk 
from tkinter import ttk
from tkinter import filedialog
import face_recognition
import cv2
from PIL import Image, ImageTk
import time 
import speech_recognition as sr
from textblob import TextBlob
```

```python
class Face_Audio_recognition_app:
    #application
    def __init__(self,root) :
        self.root = root
        self.root.title("Face And Audio Recognition App")
        
        self.face_encode = []
        self.names = []
        self.first_window()
    #first_window 
    def first_window(self):
        frame = ttk.Frame(self.root)
        train = ttk.Button(frame,text="Train The Model",command=self.train)
        capture = ttk.Button(frame,text="Take A Photo",command =self.capturing)
        display = ttk.Button(frame,text="Display The Photo",command=self.display_window)
        check = ttk.Button(frame,text="Check Identity",command=self.check_identity)
        frame.pack()
        train.pack(side=tk.LEFT, padx=5)
        capture.pack(side=tk.LEFT, padx=5)
        display.pack(side=tk.LEFT, padx=5)
        check.pack(side=tk.LEFT, padx=5)
        self.root.mainloop()
    #tarining the model 
    def train(self) :
        file_path = filedialog.askopenfilename(title="Select A photo ")
        if file_path :
            name = input("Enter the name of the person : ")
            image = face_recognition.load_image_file(file_path)
            encode = face_recognition.face_encodings(image)
            
            if encode:
                # Append the first face encoding if available
                self.face_encode.append(encode[0])
                self.names.append(name)
                print("Model trained successfully.")
            else:
                print("No face found in the provided image. Please try again.")
    #Capturing photo 
    def capturing(self):
        print("Get position")
        time.sleep(1)
        print("3")
        time.sleep(1)
        print("2")
        time.sleep(1)
        print("1")
        time.sleep(1)
        print("Smile :)")
        cap = cv2.VideoCapture(0)
        bol,frame = cap.read()
        if bol :
            cv2.imwrite("Captured image.png",frame)
            print("Image Captured Successfuly")
        else :
            print("Error Capturing Image !!")
        cap.release()
        
    #display window 
    def display_window(self) :
        images_window = tk.Toplevel(self.root)
        images_window.title("Images Display")
        self.display_images(images_window)
        back_button = ttk.Button(images_window,text="Back",command=images_window.destroy)
        back_button.pack()
    #display images 
    def display_images(self,window) :
        frame = ttk.Frame(window)
        frame.pack()
        captured_image_path = "Captured image.png"
        captured_image = Image.open(captured_image_path)
        captured_image = ImageTk.PhotoImage(captured_image)
        captured_label = ttk.Label(frame,image=captured_image)
        captured_label.image = captured_image
        captured_label.pack()

    #check_identity
    def check_identity(self):
        captured = face_recognition.load_image_file("Captured image.png")
        captured_encode = face_recognition.face_encodings(captured)[0]
        match = face_recognition.compare_faces(self.face_encode,captured_encode)
        if match[0] != False:
            name=self.names[0]
            print(f"Accepted , Welcome {name}")
            self.second_window()
        else :
            name="Unknown"
            print("Unaccepted,Try Again !! ")
    #SEcond window
    def second_window(self):
        self.clear_window()
        new_frame = ttk.Frame(self.root)
        new_frame.pack()
        speech_button = ttk.Button(new_frame,text="Speech",command=self.speech_recognition)
        speech_button.pack(side=tk.LEFT,padx=10)
        sentiment_button = ttk.Button(new_frame,text="Sentiment Analysis",command=self.sentiment_analysis)
        sentiment_button.pack(side=tk.LEFT,padx=10)
        
    # clear_window
    def clear_window(self):
        # Function to clear the window content
        for widget in self.root.winfo_children():
            widget.destroy()

    #Speech_Recognition
    def speech_recognition(self) :
        recognizer = sr.Recognizer()
        with sr.Microphone() as source :
            print("Say Something ...")
            audio = recognizer.listen(source)
        try:
            self.text = recognizer.recognize_google(audio)
            print("You Said : ",self.text)
        except :
            print("Try Again !!")
    #Sentiment Analysis
    def sentiment_analysis(self):
        blob = TextBlob(self.text)
        value = blob.sentiment.polarity
        if value > 0 :
            print("Positive Sentiment")
        elif value < 0 :
            print("Negative Sentiment")
        else : 
            print("Neutral")
        self.root.destroy()
```

```python
Face_Audio_recognition_app(tk.Tk())
```
